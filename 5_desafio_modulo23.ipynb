```python
import plotly.express as px
import pandas as pd
from dash import Dash, html, dcc
```

```python
df = pd.read_csv('C:/Users/Doris/Downloads/ecommerce_estatistica.csv')
print(df.head().to_string())
```

```python
def cria_graficos(df):
    # Histograma
    fig1 = px.histogram(df, x='Nota', nbins=30, title='DistribuiÃ§Ã£o das Notas dos Produtos')

    # GrÃ¡fico de pizza
    fig2 = px.pie(df, names='GÃªnero', color='GÃªnero', title='ParticipaÃ§Ã£o dos GÃªneros nas Vendas', color_discrete_sequence=px.colors.sequential.RdBu)

    # GrÃ¡fico de bolha
    fig3 = px.scatter(df, x='Nota', y='N_AvaliaÃ§Ãµes', size='Qtd_Vendidos_Cod', color='Desconto', hover_name='PreÃ§o', size_max=60)
    fig3.update_layout(title='NÃºmero de AvaliaÃ§Ãµes por nota e quantidade de produtos vendido')              # funÃ§Ã£o hover: mostra o campo qdo o mouse Ã© parado em cima

    # GrÃ¡fico de Linha
    fig4 = px.line(df, x='PreÃ§o', y='Qtd_Vendidos_Cod', color='N_AvaliaÃ§Ãµes', facet_col='Nota')
    fig4.update_layout(
        title='Qtd. Vendidos por PreÃ§o e NÃºmero de avaliaÃ§Ãµes para cada Nota',
        xaxis_title='PreÃ§o',
        yaxis_title='Qtd. Vendidos')

    # GrÃ¡fico #3D
    fig5 = px.scatter_3d(df, x='PreÃ§o', y='Qtd_Vendidos_Cod', z='N_AvaliaÃ§Ãµes', color='Nota')

    # GrÃ¡fico de Barra
    fig6 = px.bar(df, x='Marca', y='Qtd_Vendidos_Cod', color='GÃªnero', barmode='group', color_discrete_sequence=px.colors.qualitative.Bold, opacity=1)
    fig6.update_layout(
        title='Quantidade de Vendidos por Marca',
        xaxis_title='Marca',
        yaxis_title='Quantidade de Vendidos',
        plot_bgcolor='rgba(222, 255, 253. 1)',   # Fundo interno
        paper_bgcolor='rgba(186, 245, 241, 1)'   # Fundo externo
    )
    return fig1, fig2, fig3, fig4, fig5, fig6
```

```python
def cria_app(df):
    # cria APP
    app = Dash(__name__)
    fig1, fig2, fig3, fig4, fig5, fig6 = cria_graficos(df)
    app.layout = html.Div([
        dcc.Graph(figure=fig1),
        dcc.Graph(figure=fig2),
        dcc.Graph(figure=fig3),
        dcc.Graph(figure=fig4),
        dcc.Graph(figure=fig5),
        dcc.Graph(figure=fig6)
    ])
    return app
```

Executa App

```python
if __name__ == '__main__':
    app = cria_app(df)
    app.run(debug=True, port=8050)
```
